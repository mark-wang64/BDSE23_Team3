```python
import pandas as pd
from glob import glob
import os

def Hometown_Clean(x,y):
    US_label=["Alabama","Alaska","Arizona","Arkansas","California","Colorado","Connecticut",
        "Delaware","Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas",
          "Kentucky","Louisiana","Maine","Maryland","Massachusetts","Michigan","Minnesota",
        "Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey",
          "New Mexico","New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon",
        "Pennsylvania","Rhode Island","South Carolina","South Dakota","Tennessee","Texas",
        "Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming"]
    Taiwan_label=["Taipei","Hsinchu"]
    for i in x:
        if pd.isna(i) == False:
            mid=i.split(",")[-1]
            if mid.lstrip() in US_label:
                y.replace(i, "United States", inplace=True)
            elif mid.lstrip() == "Falkland Islands":
                y.replace(i, "United Kingdom", inplace=True)
            elif mid.lstrip() == "Centre de Flacq":
                y.replace(i, "Mauritius", inplace=True)
            else:
                y.replace(i, mid.lstrip(), inplace=True)
    for i, index in enumerate(x):
        if index is None:
            continue
        else:
            if len(index) < 3:
                y.drop(i,inplace=True)

#C:\Users\mark\Desktop\Lithuania                
path=input()
file_list=glob(path+r"\*\*.csv")
appended_data = []
for i in file_list:
    city=i.split("\\")[-2]
    size = os.path.getsize(i)
    if size > 948:
        data = pd.read_csv(i,dtype=str,on_bad_lines='skip',sep=",",index_col=None,encoding="utf-8")
        data["city"]=city
        appended_data.append(data)
    # store DataFrame in list

# see pd.concat documentation for more info
appended_data = pd.concat(appended_data,ignore_index=True)
column=["Hotel_Name","Average_Rating","Review_Date","Stay_Date","Trip_Type","Language","Room_Description","Reviw_Title"
         ,"Total_Rating","Reivew_Text","Reviewer_Name","Reviewer_Region","Value","Service","Sleep Quality","Room","Location","Cleanliness","Hotel_City","front desk","internet access"]
appended_data.set_axis(column, axis='columns', inplace=True)
df2=appended_data.where(appended_data.notnull(),None)
print("finish")
Hometown_Clean(df2["Reviewer_Region"],df2)
country=path.split("\\")[-1]
df2["Country"]=country
df2.to_csv(country+".csv",sep=",",index=None, encoding="utf-8")
```
